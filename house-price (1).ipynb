# House Price Prediction

**Import csv files**

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

**Test data**

```python
test_data=pd.read_csv("/kaggle/input/house-prices-advanced-regression-techniques/test.csv")
test_data.head()
```

**Test id for Submission **

```python
test_id=test_data["Id"].copy()
test_id.head()
```

**Train data**

```python
train_data=pd.read_csv("/kaggle/input/house-prices-advanced-regression-techniques/train.csv")
train_data.head()
```

**Data Shape**

```python
print(train_data.shape)
print(test_data.shape)
```

**Describe**

```python
train_data.describe()
```

**Null values check**

```python
train_data.isnull().sum()
```

```python
test_data.isnull().sum()
```

**Drop unwanted columns**

```python
drop_col=["Id","Alley","MasVnrType","FireplaceQu","PoolQC","Fence","MiscFeature","Utilities","Street","LandSlope",
          "Condition2","RoofMatl","LowQualFinSF","3SsnPorch","BsmtFinSF2","BsmtFinType2","BsmtHalfBath","KitchenAbvGr",
          "Functional","MoSold","YrSold"]
```

```python
train_data=train_data.drop(columns=drop_col,axis=1)
test_data=test_data.drop(columns=drop_col,axis=1)
print(train_data.shape)
print(test_data.shape)
```

```python
train_data=train_data.iloc[:-1,:].reset_index(drop=True)
```

**Features--- Input values**

```python
x=train_data.drop("SalePrice",axis=1)
print(x.shape)
print(test_data.shape)
```

**Target -- Output values**

```python
y=train_data["SalePrice"]
y.shape
```

**Numerical null values**

```python
null_num_col=["LotFrontage","MasVnrArea","BsmtFinSF1","GarageYrBlt"]
x[null_num_col].head()
```

**Category null values**

```python
null_cat_col=["Electrical","BsmtQual","BsmtCond","BsmtExposure","BsmtFinType1","GarageType","GarageFinish","GarageQual","GarageCond"]
x[null_cat_col].value_counts()
```

**Simple Imputer to fill null values**

```python
from sklearn.impute import SimpleImputer
median=SimpleImputer(strategy="median")
mode=SimpleImputer(strategy="most_frequent")
```

```python
x[null_num_col]=median.fit_transform(x[null_num_col])
x[null_cat_col]=mode.fit_transform(x[null_cat_col])
test_data[null_num_col]=median.fit_transform(test_data[null_num_col])
test_data[null_cat_col]=mode.fit_transform(test_data[null_cat_col])
```

```python
print(x[null_num_col].isnull().sum())
print(x[null_cat_col].isnull().sum())
print(test_data[null_num_col].isnull().sum())
print(test_data[null_cat_col].isnull().sum())
```

```python
x.info()
```

**Select numerical and categorical values **

```python
cat_col=x.select_dtypes(include="object").columns.tolist()
num_col=x.select_dtypes(include="number").columns.tolist()
test_cat=test_data.select_dtypes(include="object").columns.tolist()
test_num=test_data.select_dtypes(include="number").columns.tolist()
```

```python
print(cat_col)
print(test_cat)
```

```python
print(num_col)
print(test_num)
```

**Encoding the categorical values**

```python
x_encode=pd.get_dummies(x[cat_col],drop_first=True).astype("int")
test_encode=pd.get_dummies(test_data[test_cat],drop_first=True).astype("int")
```

```python
test_encode,x_encode=test_encode.align(x_encode,join="right",axis=1,fill_value=0)
```

```python
print(test_encode.shape)
print(x_encode.shape)
```

```python
print(x_encode.head(2))
print(test_encode.head(2))
```

**Features scalling**

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x_scale=pd.DataFrame(scale.fit_transform(x[num_col]),columns=num_col)
test_scale=pd.DataFrame(scale.fit_transform(test_data[test_num]),columns=test_num)
```

```python
print(x_scale[:1])
print(test_scale[:1])
```

**Combine all encoding and scalling datas**

```python
x_tdata=pd.concat([x_scale.reset_index(drop=True),x_encode.reset_index(drop=True)],axis=1)
test_tdata=pd.concat([test_scale.reset_index(drop=True),test_encode.reset_index(drop=True)],axis=1)
```

****Import Tensorflow keras ****

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

**Dense Neural Network**

```python
model=tf.keras.Sequential([
     tf.keras.layers.Dense(128,activation="relu",input_shape=(x_tdata.shape[1],)),
     tf.keras.layers.Dense(64,activation="relu"),
     tf.keras.layers.Dense(1)])
```

```python
model.compile(optimizer="adam",loss="mse",metrics=["mae"])
```

**Model fit**

```python
model.fit(x_tdata,y,epochs=100,batch_size=30)
```

**Model prediction**

```python
prediction=model.predict(test_tdata)
```

```python
prediction[:5]
```

**Create DataFrame for submission csv file**

```python
df=pd.DataFrame({"Id":test_id,"SalePrice":prediction.ravel()})
df.to_csv("Houseprice.csv",index=False)
```
